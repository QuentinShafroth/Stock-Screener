```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

import pandas as pd
tickers = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]['Symbol'].tolist()

def calculate_momentum(ticker):
    now = datetime.utcnow()
    data = yf.download(ticker, period='1y', interval='1h', progress=False)

    if data.empty:
        return None

    data = data['Close'].dropna()

    def get_return(period_hours):
        try:
            past_price = data[-int(period_hours)]['Close']
            return (data.iloc[-1] - past_price) / past_price
        except:
            return None

    try:
        momentum = {
            'Ticker': ticker,
            '1H': get_return(1),
            '1W': get_return(24*7),
            '1M': get_return(24*30),
            '1Y': get_return(24*365),
        }
        return momentum
    except:
        return None

momentum_list = []
for ticker in tickers:
    print(f"Processing {ticker}")
    m = calculate_momentum(ticker)
    if m:
        momentum_list.append(m)

df = pd.DataFrame(momentum_list)
df = df.dropna().sort_values('1W', ascending=False)  # Sort by 1-week momentum
df.to_csv("weekly_momentum.csv", index=False)
df.head()
```
